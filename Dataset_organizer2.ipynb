```python
import os
import json

# Definisci il percorso della cartella contenente i file JSON
folder_path = r'C:\Users\loren\Downloads\APKLOT-master\APKLOT-master\1. Satellite\Dataset\labelme_20 - Copia'  # Cambia questo con il percorso della tua cartella

# Funzione per rimuovere il contenuto dopo "fillColor": [255, 0, 0, 128] e aggiungere "}"
def process_json_files(folder_path):
    # Scansiona tutti i file nella cartella
    for filename in os.listdir(folder_path):
        if filename.endswith('.json'):
            file_path = os.path.join(folder_path, filename)
            # Leggi il contenuto del file JSON
            with open(file_path, 'r') as file:
                data = json.load(file)
            # Variabile per controllare se "fillColor" è stato trovato
            found_fill_color = False
            # Crea una nuova struttura dati per contenere il file JSON troncato
            truncated_data = {}
            # Itera su tutte le chiavi nel file JSON
            for key, value in data.items():
                # Se trovi "fillColor", imposta il flag a True e aggiungi questa coppia chiave-valore a truncated_data
                if key == "fillColor" and value == [255, 0, 0, 128]:
                    found_fill_color = True
                    truncated_data[key] = value
                    break
                # Altrimenti, aggiungi questa coppia chiave-valore a truncated_data
                truncated_data[key] = value
            # Aggiungi "}" alla fine del file
            truncated_data["}"] = None
            # Sovrascrivi il file con il nuovo contenuto
            with open(file_path, 'w') as file:
                json.dump(truncated_data, file, indent=4)

# Esegui la funzione per elaborare i file JSON nella cartella
process_json_files(folder_path)
```

```python
import os
import json

def remove_invalid_attribute(folder_path):
    # Scansiona tutti i file nella cartella
    for filename in os.listdir(folder_path):
        if filename.endswith('.json'):
            file_path = os.path.join(folder_path, filename)
            # Leggi il contenuto del file JSON
            with open(file_path, 'r') as file:
                data = json.load(file)
            # Rimuovi l'intera riga contenente l'attributo '}"'
            if "}" in data:
                del data["}"]
            # Sovrascrivi il file con il nuovo contenuto
            with open(file_path, 'w') as file:
                json.dump(data, file, indent=4)

# Definisci il percorso della cartella contenente i file JSON
folder_path = r'C:\Users\loren\Downloads\APKLOT-master\APKLOT-master\1. Satellite\Dataset\parcheggi\dataloader'  # Sostituisci con il percorso della tua cartella
# Esegui la funzione per rimuovere l'attributo non valido da ogni file JSON nella cartella
remove_invalid_attribute(folder_path)
```

```python
import os

def rename_files(directory):
    # Scorrere tutti i file nella directory
    for filename in os.listdir(directory):
        # Verificare se il nome del file contiene "_testing" o "_training"
        if "_testing" in filename or "_training" in filename:
            # Rinominare il file rimuovendo "_testing" o "_training" dal nome
            new_filename = filename.replace("_testing", "").replace("_training", "")
            os.rename(os.path.join(directory, filename), os.path.join(directory, new_filename))
            print(f"Rinominato il file {filename} in {new_filename}")

# Directory in cui rinominare i file
directory_path = r"C:\Users\loren\Downloads\APKLOT-master\APKLOT-master\1. Satellite\Dataset\parcheggi\dataloader"

# Chiamata alla funzione per rinominare i file nella directory specificata
rename_files(directory_path)
```

```python
import os
import shutil

def sposta_file_con_parola(directory_input, parola, directory_output):
    # Assicurati che la directory di output esista, altrimenti creala
    if not os.path.exists(directory_output):
        os.makedirs(directory_output)

    # Itera attraverso tutti i file nella directory di input
    for filename in os.listdir(directory_input):
        filepath = os.path.join(directory_input, filename)
        # Controlla se il file contiene la parola specificata
        if os.path.isfile(filepath) and parola in filename:
            # Sposta il file nella nuova directory
            shutil.move(filepath, os.path.join(directory_output, filename))
            print(f"File {filename} spostato con successo.")

# Imposta le directory di input e output
directory_input = r"C:\Users\loren\Downloads\APKLOT-master\APKLOT-master\1. Satellite\Dataset\parcheggi\segmented_images"
directory_output = r"C:\Users\loren\Downloads\APKLOT-master\APKLOT-master\1. Satellite\Dataset\parcheggi\image_newmask"

# Parola da cercare nei nomi dei file
parola = "Class"

# Chiama la funzione per spostare i file
sposta_file_con_parola(directory_input, parola, directory_output)
```
